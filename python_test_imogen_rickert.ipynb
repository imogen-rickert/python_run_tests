```python
# import required libraries
import pandas as pd
import numpy as np
from datetime import date
from scipy.stats import normaltest
```

```python
# import test dataset 
# included multiple columns to test if my function is working as expected:
# column "date" indicates 3 consecutive days of data
# column "uid" is contains unique ids
# column "id" contains a duplicate for the latest date
# column "values_norm_dist" contains values with a normal distribution
# column "values_not_norm_dist" contains values without a normal distribution
df = pd.read_csv('test_dataset.csv')
df
```

```python
# filter to only include rows with latest date
latest_df = df[df.date == max(df.date)]

# function to run both tests on imported csv
def test():
    
    # check if PK is duplicated
    # id should be duplicated, uid should be unique
    if df['uid'].duplicated().any():
        print('Error: primary key is duplicated')
    else:
        print('Primary key is unique')

    # check the distribution using D’Agostino’s K^2 Test
    # values_not_norm_dist should not appear normal, values_norm_dist should be normal
    stat, p = normaltest(df['values_not_norm_dist'])
    print('Statistics = %.3f, p = %.3f' %(stat, p))
    alpha = 0.05
    if p > alpha:
        print('The distribution appears normal')
    else:
        print('Error: the distribution does not appear normal')

test()
```

```python
# filter to only include rows with latest date
latest_df = df[df.date == max(df.date)]


def test():
    
    # check if PK is duplicated
    if latest_df['id'].duplicated().any():
        print('Error: primary key is duplicated')
    else:
        print('Primary key is unique')

    # check the distribution using D’Agostino’s K^2 Test
    stat, p = normaltest(latest_df['values_norm_dist'])
    print('Statistics = %.3f, p = %.3f' %(stat, p))
    alpha = 0.05
    if p > alpha:
        print('The distribution appears normal')
    else:
        print('Error: the distribution does not appear normal')

test()
```
